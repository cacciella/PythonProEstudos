```python
df2 = df[(df['Sales'] >= Q1) & (df['Sales'] <= Q3)]
```

##Distribuição das vendas

```python
plt.figure(figsize=(12,5))

sns.boxplot(
    x=df2["Sales"],
    color="steelblue"
)

plt.title("Média das Vendas")
plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.histplot(
    df2["Sales"],
    bins=50,
    kde=True,
    color="royalblue"
)

plt.title("Distribuição das Vendas")
plt.xlabel("Vendas")
plt.ylabel("Frequência")

plt.show()
```

```python
corr = df2.select_dtypes(include=np.number).corr()

corr
```

##Correlação

```python
plt.figure(figsize=(7.5,6))

sns.heatmap(
    corr,
    annot=True,
    cmap="RdYlBu",
    linewidths=.8,
    fmt=".2f"
)

plt.title("Mapa de Correlação")
plt.show()
```

##Produto campeão de vendas

```python
produto = (df2
           .groupby("Product Name")["Sales"]
           .sum()
           .sort_values(ascending=False)
           .head(10))
```

```python
plt.figure(figsize=(12,7))

sns.barplot(
    x=produto.values,
    y=produto.index,
    palette="viridis"
)

plt.title("Top 10 Produtos por Vendas")

plt.xlabel("Vendas")

plt.ylabel("Produto")

plt.show()
```

##País mais lucrativo

```python
pais = (df2
        .groupby("Country")["Profit"]
        .sum()
        .sort_values(ascending=False)
        .head(10))
```

```python
plt.figure(figsize=(12,7))

sns.barplot(
    x=pais.values,
    y=pais.index,
    palette="magma"
)

plt.title("Top Países por Lucro")

plt.xlabel("Lucro")

plt.ylabel("País")

plt.show()
```

##Mês mais forte

```python
df2["Order Date"] = pd.to_datetime(df2["Order Date"])

df2["Month"] = df2["Order Date"].dt.month_name()
```

```python
mes = (df2
       .groupby("Month")["Sales"]
       .sum()
       .reindex([
           'January','February','March','April','May',
           'June','July','August','September',
           'October','November','December'
       ]))
```

```python
plt.figure(figsize=(14,6))

sns.lineplot(
    x=mes.index,
    y=mes.values,
    marker="o",
    linewidth=3
)

plt.xticks(rotation=45)

plt.title("Faturamento por Mês")

plt.xlabel("Mês")

plt.ylabel("Sales")

plt.show()
```

##Desconto x Lucro

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=df,
    x="Discount",
    y="Profit",
    alpha=.6
)

plt.title("Desconto x Lucro")

plt.show()
```

Este gráfico evidencia visualmente que maiores descontos tendem a reduzir a lucratividade.

##Lucro por Categoria

```python
categoria = df.groupby("Category")["Profit"].sum()
```

```python
plt.figure(figsize=(8,5))

categoria.plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.ylabel("")
plt.title("Participação do Lucro por Categoria")

plt.show()
```

##Top Clientes

```python
clientes = (df2
            .groupby("Customer Name")["Sales"]
            .sum()
            .sort_values(ascending=False)
            .head(10))
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    x=clientes.values,
    y=clientes.index,
    palette="crest"
)

plt.title("Top 10 Clientes")

plt.show()
```
