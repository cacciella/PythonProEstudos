```python
import seaborn as sns
```

```python
Q1 = df["Sales"].quantile(0.25)
Q3 = df["Sales"].quantile(0.75)

IQR = Q3 - Q1

print(IQR)
```

```python
df2 = df[(df['Sales'] >= Q1) & (df['Sales'] <= Q3)]
```

```python
plt.figure(figsize=(12,6))

sns.histplot(
    df2["Sales"],
    bins=50,
    kde=True,
    color="green"
)

plt.title("Distribuição das Vendas")
plt.xlabel("Sales")
plt.ylabel("Frequência")

plt.show()
```

```python

df.plot(x='Market',y='Sales')
```

```python
plt.figure(figsize=(9,4))

sns.boxplot(
    x=df2["Sales"],
    color="green"
)

plt.title("Média das Vendas")
plt.show()
```

###Outros Modelos

```python
import matplotlib.pyplot as plt

x = [1, 2, 3]

y_1 = [50, 60, 70]
y_2 = [20, 30, 40]

plt.plot(x, y_1, marker='x')
plt.plot(x, y_2, marker='^')

plt.xlim([0, len(x)+1])
plt.ylim([0, max(y_1+y_2) + 10])
plt.xlabel('Label X')
plt.ylabel('Label Y')
plt.title('Gráfico de linhas simples')
plt.legend(['amostra 1', 'amostra 2'], loc='upper left')

plt.show()
```
